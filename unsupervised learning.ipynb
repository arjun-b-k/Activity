# STANDARD SCALING

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scaled_data = scalar.fit_transform(data)
```

# K MEANS CLUSTERING

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
ks=range(1,11)
inertia=[]
for k in ks:
    model=KMeans(n_clusters=k,init="k-means++",random_state=42)
    model.fit(scaled_data)
    inertia.append(model.inertia_)
print(inertia)
```

```python
plt.plot(ks,inertia,'-o')
plt.title("Elbow Method Plot")
plt.xlabel('No of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
model_kmeans=KMeans(n_clusters=3)
model_kmeans.fit(scaled_data)
km=model_kmeans.predict(scaled_data)
print(model_kmeans.labels_)
```

# HIERARCHIAL CLUSTERING

```python
from scipy.cluster.hierarchy import linkage,dendrogram

dendro = linkage(scaled_data,method = 'ward')
dendrogram(dendro, leaf_rotation = 90, leaf_font_size = 6)
plt.hlines(20,0,3000,colors="red",linestyles="dashed")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="ward")
aghc=cluster.fit_predict(scaled_data)
aghc
```
